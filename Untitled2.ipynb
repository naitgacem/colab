```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
from random import randint

def generate_sine(freq :int,\
                  amplitude : int , \
                  sampling_freq : int, \
                  duration:int) \
                  -> tuple:

    sampling_interval = 1 / sampling_freq
    
    t_vec = np.linspace(-duration/2, duration/2, int(duration/sampling_interval), endpoint = False)
    y_vec = amplitude * np.sin(2 * np.pi * freq* t_vec)
    return t_vec, y_vec

def sum_sines(*waves : tuple) -> tuple :
  sum = np.zeros(waves[0][1].shape)
  time = waves[0][0]
  for wave in waves :
    
    sum += wave[1]
  return time, sum 

sampling_freq = 40000 #4kHz
amp1 = 4
amp2 = 2
duration = 2#seconds

slow_freq = 1#Hz
fastest_freq = 10#Hz

signal = generate_sine(slow_freq, amp1, sampling_freq, duration)
plt.figure(0)

input_str = input("Enter a list of frequencies delimited by space  > ")
input_list = input_str.split(' ')

list_of_freq = []
for freq in input_list :
    list_of_freq.append(int(freq))
#list_of_freq =   #[2, 5, 10]

for freq in list_of_freq :
  plt.plot(*(signal))
  amplitude = randint(1, 7)
  signal = sum_sines(signal, generate_sine(freq,amplitude, sampling_freq, duration ))
  plt.plot(*(signal))

plt.figure(1)
sum_of_sines = signal
plt.plot(*(sum_of_sines))

N = len(sum_of_sines[1])
yf = 2 * scipy.fft.rfft(sum_of_sines[1])/N
yf[0] /= 2 
xf = scipy.fft.rfftfreq(int(N), 1/sampling_freq)



plt.figure(2)
plt.plot(xf, np.abs(yf))
plt.xlim(-1,40)
```
